```python
city = ["chennai", "bangalore", "delhi", "mumbai"]

# Replace the 3rd element
city[2] = "punjab"

# Append correctly
city.append("delhi")

# Display the whole list
print(f"Cities: {', '.join(city)}")
```

```python
city.insert(3,"maturai")
print(f"city: " ,city)
```

```python
city.extend(["ooty","coimbatore"])
print(f"city: " ,city)
```

```python
city.remove("mumbai")
print(f"city: " ,city)
```

```python
city.pop()
print(f"city: " ,city)
```

```python
city.pop(2)
print(f"city: " ,city)
```

```python
city.sort()
print(f"city: " ,city)
```

```python
city.reverse()
print(f"city: " ,city)
```

```python

city.sort(reverse=True)
print(f"city: " ,city)
```

```python
## Example of sort & Reverse
Num = [10,20,30,40,50,60]
Num.sort()
print(f"Sorting Order: {Num}")

Num.reverse()
print(f"Reverse Order: {Num}")

Num.sort(reverse=True)
print(f"Reverse Order: {Num}")
```

```python
Num = [10, 20, 30, 40, 50, 60]

# Indexing
print(Num[0])
print(Num[-1])

#  Slicing
print(Num[1:4])
print(Num[:0:-1])
```
