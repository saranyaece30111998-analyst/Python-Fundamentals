Case Study – Email Username Generator Instructions:

• Create variables: full_name, domain_name

• Take input from the user using input() function

• Convert full_name into lowercase using lower()

• Remove spaces in name and replace with dot using replace()

• Split full_name using split() function

• Join name parts using join() function if needed

• Create email in format: name@domain

• Display only Full Name and Generated Email

• Use print() and f-string for output display


```python
# creation of varibles
full_name = input("Enter your full name: ")
domain_name = input("Enter your domain name: ")

# Convert  lowercase
full_name = full_name.lower()

# Replace spaces with dots
full_name = full_name.replace(" ", ".")

# Split by dot
name_new = full_name.split(".")

# Join  with dot
name2 = ".".join(name_new)

# Generate email
email = f"{name2}@{domain_name}"

# Display results
print(f"Full Name: {full_name}")
print(f"Generated Email: {email}")
```
