```python
# atm pin verification system

# step 1: create a variable with default pin
correct_pin = "1234"   # you can set any pin here

# step 2: use a while loop to repeatedly ask for pin
while True:
    entered_pin = input("enter your atm pin: ")

    # step 3: check if entered pin matches
    if entered_pin == correct_pin:
        print("login successful")
        break   # exit the loop if pin is correct
    else:
        print("incorrect pin... try again")

```

Case Study – Password Verification System with Limited Attempts
Instructions:

Create variables: • password
• attempts
• max_attempts

Use a while loop to allow the user to enter the password.

Check whether the entered password matches the original password.

If the password is correct: • Display "Access granted!"
• Stop the loop using break

If the password is incorrect: • Increase attempt count
• Display remaining attempts

If maximum attempts are reached: • Display "Maximum attempts reached. Access denied!"

```python
# password verification system with limited attempts

password = "secret123"   # set your original password here
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    entered_password = input("enter your password: ")

    if entered_password == password:
        print("access granted!")
        break
    else:
        attempts += 1
        remaining = max_attempts - attempts
        if remaining > 0:
            print(f"incorrect password. {remaining} attempts left.")
        else:
            print("maximum attempts reached. access denied!")
```
