```python
# 1. Create a dictionary named student with keys: name, age, course
student = {
    "name": ["Saranya", "John", "Priya"],
    "age": [27, 22, 24],
    "course": ["Data Analyst", "Python Developer", "AI Engineer"]
}

# 2. Display initial student details
print(f"Initial Student Details:{student}")


# 3. Add one new student details
student["name"].append("Kumar")
student["age"].append(23)
student["course"].append("Web Developer")

# 4. Display all student details using dictionary keys
print(f"\nUpdated Student Details:")
print(f"Names:{student["name"]}")
print(f"Ages:{ student["age"]}")
print(f"Courses:{ student["course"]}")

# 5. Access and display:
# First student's details

print("\nFirst Student Details:")
print(f"Name: {student['name'][0]}")
print(f"Age: {student['age'][0]}")
print(f"Course: {student['course'][0]}")

# Newly added student's details (last index)
print("\nNewly Added Student Details:")
print(f"Name:{student["name"][-1]}")
print(f"Age:{student["age"][-1]}")
print(f"Course: {student["course"][-1]}")
```

Case Study – Student Course Registration System using Set
Instructions:

Create a set named courses with course names: Python, SQL, Power BI, Python, Excel

Perform set operations: • Add course using add()
• Add multiple courses using update()
• Remove course using remove() and discard()

Display all courses and total number of courses using len()


```python
course = {"python", "sql", "powerbi", "excel"}
print(f"original course: {course}")

# Add a single item
course.add("tableau")
print(f"after add: {course}")

# Update with multiple items (use a list, set, or tuple)
course.update(["html", "ai"])
print(f"after update: {course}")


print(f"after remove: {course}")

# Remove an item using discard()
course.discard("excel")
print(f"after discard: {course}")

# Display all courses
print(f"all courses: {course}")

# Display total number of courses
print(f"total number of courses: {len(course)}")


```
