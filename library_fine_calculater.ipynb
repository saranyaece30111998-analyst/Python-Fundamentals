Case Study – Library Fine Calculator
Instructions:

Create variables:

student_name

book_name

days_late

fine_per_day

Use input() function to get details from the user.

Use type casting:

int() for days_late

float() for fine_per_day

Calculate:

total_fine = days_late * fine_per_day

Display all details using:

print()

f-string

Display datatype of important variables using:

type()



```python
#creating variables
student_name = input("Enter student name: ")
book_name = input("Enter book name: ")
days_late = int(input("Enter number of days late: "))
fine_per_day = float(input("Enter fine per day: "))

# calculate total fine
total_fine = days_late * fine_per_day

# displaying details
print("\n----- Library Fine Details -----")
print(f"Student Name   : {student_name}")
print(f"Book Name      : {book_name}")
print(f"Days Late      : {days_late}")
print(f"Fine per Day   : {fine_per_day}")
print(f"Total Fine     : {total_fine}")

# displaying data types
print("\n----- Displaying Data Types -----")
print(f"Type of student_name : {type(student_name)}")
print(f"Type of book_name    : {type(book_name)}")
print(f"Type of days_late    : {type(days_late)}")
print(f"Type of fine_per_day : {type(fine_per_day)}")
print(f"Type of total_fine   : {type(total_fine)}")
```
